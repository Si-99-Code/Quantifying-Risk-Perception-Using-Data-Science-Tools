```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
```

```python
df_graph = pd.read_csv('merged-websites.csv', names = ['Node 1', 'Node 2'])
G = nx.Graph()
```

```python
G = nx.from_pandas_edgelist(df_graph, 'Node 1', 'Node 2')
```

```python
degree_freq = nx.degree_histogram(G)
degrees = range(len(degree_freq))
plt.figure(figsize=(12, 8)) 
plt.loglog(degrees, degree_freq,'go-') 
plt.title('Degree and Frequency of all websites combined')
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.show()
```

```python
degree_sequence = sorted((d for n, d in G.degree()), reverse=True)

fig = plt.figure("Degree of a random graph", figsize=(10, 15))
axgrid = fig.add_gridspec(5, 4)

ax0 = fig.add_subplot(axgrid[0:3, :])
Gcc = G.subgraph(sorted(nx.connected_components(G))[0])
pos = nx.spring_layout(Gcc)
nx.draw_networkx_nodes(Gcc, pos, node_size=10)
nx.draw_networkx_edges(Gcc, pos)
ax0.set_title("Connected components of all websites combined")
ax0.set_axis_on()
```

```python
plt.plot(degree_sequence, marker=".")
plt.title("Degree Rank Plot")
plt.ylabel("Degree")
plt.xlabel("Rank")

plt.show()
```

```python
plt.bar(*np.unique(degree_sequence, return_counts=True))
plt.title("Degree histogram")
plt.xlabel("Degree")
plt.ylabel("# of Nodes")

plt.show()
```
